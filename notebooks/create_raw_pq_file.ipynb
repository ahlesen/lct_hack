```python
class TextTransliteration:
    def __init__(self, model_name="Helsinki-NLP/opus-mt-en-ru"):
        self.model_name = model_name
        self.model = MarianMTModel.from_pretrained(model_name)
        self.tokenizer = MarianTokenizer.from_pretrained(model_name)

    def transliterate_to_russian(self, text):
        batch = self.tokenizer([text], return_tensors="pt")
        translated = self.model.generate(**batch)
        translated_text = self.tokenizer.batch_decode(translated, skip_special_tokens=True)[0]
        return translated_text
```

```python
import re
def _basic_text_from_image_preprocessing(text: str) -> str:
    text = text.lower()
    text = re.sub('[^a-z0-9,. ]+', ' ', text)
    text = re.sub('[,.]', ' ', text)
    return ' '.join([word for word in set(text.split()) if word not in english_stopwords])
```

```python
from tqdm.notebook import tqdm
tqdm.pandas()
```

```python
df["title shazam"] = df["title shazam"].progress_apply(lambda x: _basic_text_from_image_preprocessing(x).title())
df["subtitle shazam"] = df["subtitle shazam"].progress_apply(lambda x: _basic_text_from_image_preprocessing(x).title())
```

```python
title_shazam = df["title shazam"].unique()
subtitle_shazam = df["subtitle shazam"].unique()
```

```python
title_shazam_transliterated = []
for t in tqdm(title_shazam):
    title_shazam_transliterated.append(text_translit.transliterate_to_russian(t))
```

```python
df_title_shazam_transliterated = pd.DataFrame()
df_title_shazam_transliterated["title shazam"] = title_shazam
df_title_shazam_transliterated["title shazam transliterated"] = title_shazam_transliterated
df_title_shazam_transliterated.to_parquet("df_title_shazam_transliterated.pq")
```

```python
subtitle_shazam_transliterated = []
for t in tqdm(subtitle_shazam):
    subtitle_shazam_transliterated.append(text_translit.transliterate_to_russian(t))
```
